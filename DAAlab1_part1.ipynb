```python
import time
import matplotlib.pyplot as plt
import random

# Part 1: Sum of first N natural numbers

def iterative_sum(n):
    sum_iter = 0
    for i in range(1, n + 1):
        sum_iter += i
    return sum_iter

def recursive_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sum(n - 1)


# Changing n value and calculating time taken
n_values = [10, 100,]
times_iterative = []
times_recursive = []

for n in n_values:
    start_time = time.time()
    iterative_sum(n)
    end_time = time.time()
    times_iterative.append(end_time - start_time)

    start_time = time.time()
    recursive_sum(n)
    end_time = time.time()
    times_recursive.append(end_time - start_time)

# Plot the results
plt.plot(n_values, times_iterative, label='Iterative')
plt.plot(n_values, times_recursive, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time taken for calculating sum of first N natural numbers')
plt.legend()
plt.show()


# Part 2: Linear and Binary Searches

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generating random array
arr = [random.randint(1, 1000) for _ in range(10000)]
arr.sort() #sorting the array as binary search requires sorted array

# Timing linear and binary searches for 5 different search keys
search_keys = [random.randint(1, 1000) for _ in range(5)]
linear_times = []
binary_times = []

for key in search_keys:
    start_time = time.time()
    linear_search(arr, key)
    end_time = time.time()
    linear_times.append(end_time - start_time)

    start_time = time.time()
    binary_search(arr, key)
    end_time = time.time()
    binary_times.append(end_time - start_time)

# Plotting the results
plt.plot(search_keys, linear_times, label='Linear Search')
plt.plot(search_keys, binary_times, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Time (seconds)')
plt.title('Time taken for linear and binary searches')
plt.legend()
plt.show()


# Part 3: Recursive function to convert string of digits to integer

def string_to_integer(s):
    if len(s) == 0:
        return 0
    else:
        return int(s[0]) * 10 ** (len(s) - 1) + string_to_integer(s[1:])


print(string_to_integer("13531"))  


# Part 4: Recursive function to reverse a string

def reverse_string(s):
    if len(s) <= 1:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

print(reverse_string("pots&pans")) 


# Part 5: Recursive function to check if a string is palindrome

def is_palindrome(s):
    if len(s) <= 1:
        return True
    else:
        return s[0] == s[-1] and is_palindrome(s[1:-1])


print(is_palindrome("racecar"))
print(is_palindrome("gohangasalamiimalasagnahog")) 
```
